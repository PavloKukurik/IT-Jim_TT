```python
import os
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image
from collections import Counter
```

```python
train_dir = '../dataset/train'
test_dir = '../dataset/test'

train_files = os.listdir(train_dir)
test_files = os.listdir(test_dir)
```

```python
train_labels = [f.split('_')[-1].split('.')[0] for f in train_files]

label_counts = Counter(train_labels)
print("Train class distribution:", label_counts)
```

```python
sns.barplot(x=list(label_counts.keys()), y=list(label_counts.values()))
plt.title('Train Class Distribution')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
```

```python
fig, axs = plt.subplots(1, 4, figsize=(12, 3))
for idx, cls in enumerate(['0', '1']):
    examples = [f for f in train_files if f.endswith(f'_{cls}.png')][:2]
    for i, fname in enumerate(examples):
        img = Image.open(os.path.join(train_dir, fname))
        axs[idx*2 + i].imshow(img)
        axs[idx*2 + i].set_title(f'Class {cls}')
        axs[idx*2 + i].axis('off')
plt.tight_layout()
plt.show()
```
